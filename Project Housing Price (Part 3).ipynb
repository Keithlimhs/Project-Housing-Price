## Part 3: Exploratory data analysis

```python
# Step 1: Import the four libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Load CSV
df = pd.read_csv('housing_data_cleaned.csv')
```

## Univariate analysis (UA)
In this section, we will perform univariate analysis. We'll examine each column with either a histogram or a barplot. 

For categorical values, we will first get a frequent count followed by plotting of the barplot.

```python
# Step 3: Plot a barplot for "town"
plt.figure(figsize = (16,8))
df['town'].value_counts().plot.barh()
```

```python
# Step 4: Plot a barplot with "flat_type"
plt.figure(figsize = (16,8))
df['flat_type'].value_counts().plot.barh()
```

```python
# Step 5: Plot a barplot with "storey_range"
plt.figure(figsize = (16,8))
df['storey_range'].value_counts().plot.barh()
```

```python
# Step 6: Plot a histogram with "floor_area_sqm"
plt.figure(figsize=(16,8))
df['floor_area_sqm'].hist()
plt.xlabel('floor area')
plt.ylabel('Count')
```

```python
# Step 7: Plot a barplot with "flat_model"
plt.figure(figsize = (16,8))
df['flat_model'].value_counts().plot.barh()
```

```python
# Step 8: Plot a barplot with "lease_commense_date" 
plt.figure(figsize = (16,8))
df['lease_commence_date'].value_counts().plot.bar()
```

```python
# Step 9: Plot a histogram with "resale_price"
plt.figure(figsize=(16,8))
df['resale_price'].hist()
plt.xlabel('price')
plt.ylabel('Count')
```

```python
# Step 10: Plot a barplot with 'year' (you'll see interesting things)
# In fact, you can try to find out why you observe such a pattern in certain years
plt.figure(figsize = (16,8))
df['year'].value_counts().plot.bar()
```

```python
# Step 11: Plot a barplot with "real_month"
plt.figure(figsize = (16,8))
df['real_month'].value_counts().plot.bar()
```

```python
# Step 12: Plot remaining_lease using a barplot
plt.figure(figsize = (16,8))
df['remaining_lease'].value_counts().plot.bar()
```
