## Part 3: Exploratory data analysis

```python
# Step 1: Import the four libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Load CSV
df = pd.read_csv('housing_data_cleaned.csv')
df
```

## Univariate analysis (UA)
In this section, we will perform univariate analysis. We'll examine each column with either a histogram or a barplot. 

For categorical values, we will first get a frequent count followed by plotting of the barplot.

```python
# Step 3: Plot a barplot for "town"
plt.figure(figsize = (16,8))
df['town'].value_counts().plot.barh()
```

```python
# Step 4: Plot a barplot with "flat_type"
plt.figure(figsize = (16,8))
df['flat_type'].value_counts().plot.barh()
```

```python
# Step 5: Plot a barplot with "storey_range"
plt.figure(figsize = (16,8))
df['storey_range'].value_counts().plot.barh()
```

```python
# Step 6: Plot a histogram with "floor_area_sqm"
plt.figure(figsize=(16,8))
df['floor_area_sqm'].hist()
plt.xlabel('floor area')
plt.ylabel('Count')
```

```python
# Step 7: Plot a barplot with "flat_model"
plt.figure(figsize = (16,8))
df['flat_model'].value_counts().plot.barh()
```

```python
# Step 8: Plot a barplot with "lease_commense_date" 
plt.figure(figsize = (16,8))
df['lease_commence_date'].value_counts().plot.bar()
```

```python
# Step 9: Plot a histogram with "resale_price"
plt.figure(figsize=(16,8))
df['resale_price'].hist()
plt.xlabel('price')
plt.ylabel('Count')
```

```python
# Step 10: Plot a barplot with 'year' (you'll see interesting things)
# In fact, you can try to find out why you observe such a pattern in certain years
plt.figure(figsize = (16,8))
df['year'].value_counts().plot.bar()
```

```python
# Step 11: Plot a barplot with "real_month"
plt.figure(figsize = (16,8))
df['real_month'].value_counts().plot.bar()
```

```python
# Step 12: Plot remaining_lease using a barplot
plt.figure(figsize = (16,8))
df['remaining_lease'].value_counts().plot.bar()
```

## Bivariate Analysis (BA)
Next up, bivariate analysis. In univariate analysis, we looked at understanding each column by themselves. In this section, we will take a look at the relationship columns have with each other. Usually these relationships are between an indepedent variable, e.g., town, and the dependent variable, i.e. resale price. 

We'll be looking at the relationship resale price has with other columns, e.g., town, flat_type, etc. 

```python
# Step 13: Plot year vs resale_price with a boxplot
plt.figure(figsize = (16,8))
sns.boxplot(df['year'], df['resale_price'])
```

```python
# Step 14: Plot month vs resale_price with a boxplot
plt.figure(figsize = (16,8))
sns.boxplot(df['real_month'], df['resale_price'])
```

```python
# Step 15: Perform a boxplot using 'flat_type' vs 'resale_price'. Try to reorder the boxes if you can
plt.figure(figsize = (16,8))
sns.boxplot(df['flat_type'], df['resale_price'], order = ['1 ROOM', '2 ROOM','3 ROOM','4 ROOM', '5 ROOM', 'EXECUTIVE', 'MULTI GENERATION'])
```

```python
# Step 16: Perform a boxplot using "flat_model" and "resale_price"
plt.figure(figsize = (16,8))
sns.boxplot(df['flat_model'], df['resale_price'])
plt.xticks(rotation = 45)
```

```python
# Step 17: Perform a boxplot using "town" and "resale_price"
plt.figure(figsize = (16,8))
sns.boxplot(df['town'], df['resale_price'])
plt.xticks(rotation = 45)
```

```python
# Step 18: Plot floor_area_sqm vs resale_price
plt.figure(figsize = (16,8))
sns.scatterplot(df['floor_area_sqm'], df['resale_price'])
```

```python
# Step 19: Plot remaining_lease vs resale_price
plt.figure(figsize = (16,8))
sns.scatterplot(df['remaining_lease'], df['resale_price'])
```
