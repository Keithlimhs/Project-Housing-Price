## Part 5: Machine learning model and training

```python
# Step 1: Import your libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Step 2a: Load Part IV CSV
df = pd.read_csv('housing_data_cleaned_final.csv')
df
```

```python
# Step 2b: Check your columns
df.columns
```

```python
# Step 3: Drop the features that you do no need
df = df.drop(['month', 'town', 'flat_type', 'block', 'street_name', 'storey_range', 'flat_model'], axis = 1)
df
```

```python
# Step 4a: Prepare your DataFrame containing only your independent variables
X = df.drop('resale_price', axis = 1)

# Step 4b: Get the 'resale_price' column and assign it to a variable
y = df['resale_price']
```

```python
# Step 5: Import the libraries your need
from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Step 6: Split your data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# Step 7a: Declare a variable to store the model
dummy_regression = DummyRegressor()
# Step 7b: Fit your train dataset
dummy_regression.fit(X_train, y_train)
# Step 7c: Declare a variable and store your predictions that you make with your model using X test data
y_hat_dummy = dummy_regression.predict(X_test)
```

```python
# Step 8: Repeat Step 7 with other models
lr = LinearRegression()
lr.fit(X_train, y_train)
y_hat_lr = lr.predict(X_test)
```

```python
tree = DecisionTreeRegressor()
tree.fit(X_train, y_train)
y_hat_tree = tree.predict(X_test)
```

```python
forest = RandomForestRegressor()
forest.fit(X_train, y_train)
y_hat_forest = forest.set_params(n_jobs = 1).predict(X_test)
```

### Step 9: Assess your model performance
We'll be using two ways to assess our model.

1. Scatter plot comparing the actual values of the dependent variable and the predictions
2. The root mean-squared-error (RMSE) score

```python
# Step 9a: Print the RMSE between the y test and the prediction
print(np.sqrt(mean_squared_error(y_test, y_hat_dummy)))

# Step 9b: Plot a scatter plot test dependent variables vs predictions
plt.figure(figsize = (16,8))
sns.scatterplot(y_test, y_hat_dummy)
```

```python
# Step 9a: Print the RMSE between the y test and the prediction
print(np.sqrt(mean_squared_error(y_test, y_hat_lr)))

# Step 9b: Plot a scatter plot test dependent variables vs predictions
plt.figure(figsize = (16,8))
sns.scatterplot(y_test, y_hat_lr)
```

```python
# Step 9a: Print the RMSE between the y test and the prediction
print(np.sqrt(mean_squared_error(y_test, y_hat_tree)))

# Step 9b: Plot a scatter plot test dependent variables vs predictions
plt.figure(figsize = (16,8))
sns.scatterplot(y_test, y_hat_tree)
```

```python
# Step 9a: Print the RMSE between the y test and the prediction
print(np.sqrt(mean_squared_error(y_test, y_hat_forest)))

# Step 9b: Plot a scatter plot test dependent variables vs predictions
plt.figure(figsize = (16,8))
sns.scatterplot(y_test, y_hat_forest)
```

### Step 10: Get the feature importance of each of your features for your best model
Last but not least, you want to know what's going on in your model. You can do that by examining the feature importance attribute that almost all models have.

Knowing this information is useful since you can think about a data story that you can tell your stakeholders who are not technical. 

You can do for all of your models, but start with picking the model that generated predictions with the smallest RMSE, i.e. the best performing model.

```python
# Step 10a: get a list containing the columns in your train dataframe
columns = X_train.columns

# Step 10b: get the feature importances from your model
features = forest.feature_importances_
# Step 10c: declare a variable containing a new DataFrame that you create from Steps 10a and 10b
feature_importance = pd.DataFrame({'features': columns, 'importance': features})

# Step 10d: sort your newly created DataFrame by the importance score
feature_importance.sort_values(by ='importance', ascending = False, inplace = True)
# Step 10e: view your DataFrame
feature_importance
```
